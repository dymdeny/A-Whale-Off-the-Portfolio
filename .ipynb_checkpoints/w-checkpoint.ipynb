```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
 # Reading whale returns
whale_returns_csv = Path("../Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_csv,index_col="Date", parse_dates=True, infer_datetime_format=True)
whale_returns.head()
```

```python
 # Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns = whale_returns.dropna().copy()
whale_returns
```

```python
# Reading algorithmic returns
algo_returns_csv = Path("../Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_csv,index_col="Date",parse_dates =True,infer_datetime_format=True)
algo_returns.head()                        
```

```python
 # Count nulls
algo_returns.isnull().sum()
```

```python
 # Drop nulls
algo_returns=algo_returns.dropna().copy()
algo_returns.head()
```

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_history_csv = Path("../Resources/sp_tsx_history.csv")
sp_tsx_history = pd.read_csv(sp_tsx_history_csv,index_col="Date",parse_dates =True,infer_datetime_format=True)
sp_tsx_history.head()
```

```python
# Check Data Types
sp_tsx_history.dtypes
```

```python
# Fix Data Types
sp_tsx_history["Close"] =sp_tsx_history["Close"].astype(str).str.replace('$','',regex=True)
sp_tsx_history
```

```python
sp_tsx_history["Close"]=sp_tsx_history["Close"].astype(float)
sp_tsx_history.dtypes
```

```python
  # Calculate Daily Returns
sp_tsx_daily_returns = (sp_tsx_history - sp_tsx_history.shift(1)) / sp_tsx_history.shift(1)
sp_tsx_daily_returns.head()
```

```python
 # Drop nulls
sp_tsx_daily_returns.dropna(inplace = True)
sp_tsx_daily_returns.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sp_tsx_daily_returns.rename(columns = {'Close':'S&P 60 TSX'},inplace = True)
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
joined_data = pd.concat([sp_tsx_daily_returns,algo_returns,whale_returns],axis = "columns",join = "inner")
```

```python
 # Plot daily returns of all portfolios
joined_data.plot(figsize=(20,10))
```

```python
 # Calculate cumulative returns of all portfolios
cumulative_returns = (1 + joined_data).cumprod()
cumulative_returns.head()
```

```python
# Plot cumulative returns
cumulative_returns.plot(figsize=(20,10))
```

```python
#Determine the risk of each portfolio:
```

```python
# Box plot to visually show risk
#Create a box plot for each portfolio. 
joined_data.plot( kind="box",figsize =(20,10))
```

```python
 # Calculate the daily standard deviations of all portfolios
#Calculate the standard deviation for all portfolios.
joined_data_std =joined_data.std()
joined_data_std
```

```python
#Determine which portfolios are riskier than the S&P TSX 60.
riskier_std = joined_data_std [joined_data_std > joined_data_std.loc["S&P 60 TSX"]]
riskier_std
```

```python
#Calculate the Annualized Standard Deviation.
portfolios_annualized_std = joined_data.std() * np.sqrt(252)
portfolios_annualized_std 
```

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
portfolios_rolling_21=joined_data.rolling(window = 21).std()
```

```python
# Plot the rolling standard deviation
portfolios_rolling_21.plot(figsize = (20,10))
```

```python
# Calculate the correlation
# Display de correlation matrix
correlation_sp_tsx=joined_data.corr()
correlation_sp_tsx
```

```python
# Display de correlation matrix
joined_data.plot(kind="scatter", x = "BERKSHIRE HATHAWAY INC", y = "S&P 60 TSX", figsize = (20,10))
```

```python
# Calculate covariance of a single portfolio
rolling_covariance = joined_data['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(joined_data['S&P 60 TSX'])
```

```python
# Calculate variance of S&P TSX
rolling_variance = joined_data['S&P 60 TSX'].var()
rolling_variance
```

```python
# Computing beta
rolling_beta= rolling_covariance / rolling_variance
rolling_beta
```

```python
# Plot beta trend
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of BRK')
rolling_beta
```

```python
# Use `ewm` to calculate the rolling window
rolling_window_ewm = joined_data.ewm(halflife = 21,adjust=True, ignore_na=False).std().plot(title='Exponentially Weighted Moving Average',figsize=(20, 10))
rolling_window_ewm
```

```python
# Annualized Sharpe Ratios
annualized_sharp_ratios = joined_data.mean()*252/(joined_data.std()*np.sqrt(252))
annualized_sharp_ratios
```

```python
 # Visualize the sharpe ratios as a bar plot
annualized_sharp_ratios.plot(kind="bar", title="Sharpe Ratios")
```

```python
# Reading data from 1st stock
sq_history = pd.read_csv(Path("../Resources/SQ.csv"), index_col="Date",parse_dates = True, infer_datetime_format= True)
sq_history['Symbol']='Square Inc.'
sq_history = sq_history[['Symbol','Close']]
sq_history.head()
```

```python
 # Reading data from 2nd stock
zm_history = pd.read_csv(Path("../Resources/ZM.csv"), index_col="Date",parse_dates = True, infer_datetime_format= True)
zm_history['Symbol']='Zoom Video  Inc'
zm_history = zm_history[['Symbol','Close']]
zm_history.head()
```

```python
 # Reading data from 3rd stock
intc_history = pd.read_csv(Path("../Resources/INTC.csv"), index_col="Date",parse_dates = True, infer_datetime_format= True)
intc_history['Symbol']='Intel Corporation'
intc_history = intc_history[['Symbol','Close']]
intc_history.head()
```

```python
 # Combine all stocks in a single DataFrame
my_stocks=pd.concat((sq_history,zm_history,intc_history),axis = 1, join= "inner")
my_stocks
```

```python
# Calculate daily returns

my_stocks_daily_returns = (my_stocks - my_stocks.shift(1)) / my_stocks.shift(1)
my_stocks_daily_returns.dropna(inplace=True)
my_stocks_daily_returns.head()
```

```python
# Drop NAs
my_stocks.dropna()
my_stocks
```

```python
# Set weights
weights = [1/3, 1/3, 1/3]
my_portfolio= my_stocks.dot(weights)
```

```python
# Calculate portfolio return
my_portfolio_returns = my_portfolio.pct_change()
my_portfolio_returns.dropna(inplace = True)
my_portfolio_returns.head()
```

```python
 # Join your returns DataFrame to the original returns DataFrame
portfolio_df = pd.concat([whale_returns,algo_returns, sp_tsx_history, my_portfolio_returns], axis=1, join='inner')
del portfolio_df ["Close"]
portfolio_df.rename(columns = {0:"My Portfolio"}, inplace = True)
portfolio_df
```

```python
 # Only compare dates where return data exists for all the stocks (drop NaNs)
portfolio_df.dropna(inplace=True)
portfolio_df 
```
